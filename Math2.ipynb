2) Нарисуйте трехмерный график двух любых поверхностей второго порядка.

```python
def fun1(x):
    return x

def fun2(x):
    return x + 1

fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
x = y = np.arange(-3.0, 3.0, 0.05)
X, Y = np.meshgrid(x, y)
zs = np.array(fun1(np.ravel(X)))
Z = zs.reshape(X.shape)
ax.plot_surface(X, Y, Z)

x = y = np.arange(-3.0, 3.0, 0.05)
X, Y = np.meshgrid(x, y)
zs = np.array(fun2(np.ravel(X)))
Z = zs.reshape(X.shape)
ax.plot_surface(X, Y, Z)

plt.show()
```

```python
def fun1(x, y):
    return x**2 + y

def fun2(x, y):
    return x**2 - y

fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
x = y = np.arange(-3.0, 3.0, 0.05)
X, Y = np.meshgrid(x, y)
zs = np.array(fun1(np.ravel(X), np.ravel(Y)))
Z = zs.reshape(X.shape)
ax.plot_surface(X, Y, Z)

x = y = np.arange(-3.0, 3.0, 0.05)
X, Y = np.meshgrid(x, y)
zs = np.array(fun2(np.ravel(X), np.ravel(Y)))
Z = zs.reshape(X.shape)
ax.plot_surface(X, Y, Z)

plt.show()
```

```python
def polar_to_carthesian(theta, rho):
    x = rho * np.cos(theta)
    y = rho * np.sin(theta)
    return x, y
```

```python
ax = subplot(polar = True)
circle = plt.Circle((0.5, 0.3), 0.2, transform = ax.transData._b, fill = False)
ax.add_artist(circle)
plt.show()
```

### Задание 4

1) Решите систему уравнений:      
y = x^2 – 1     
exp(x) + x∙(1 – y) = 1

```python
from scipy.optimize import fsolve

def equations(p):
    x, y = p
    return (x**2 - 1 - y, np.exp(x) + x * (1 - y) - 1)

x, y =  fsolve(equations, (np.random.randint(10), np.random.randint(10)))

print (x, y)
```

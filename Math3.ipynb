1) Напишите код, проверяющий любую из теорем сложения или умножения вероятности на примере рулетки или подбрасывания монетки.  
2) Сгенерируйте десять выборок случайных чисел х0, …, х9 и постройте гистограмму распределения случайной суммы х0+х1+ …+ х9.

```python
# P(A or B) = P(A) + P(B)
# A roulette is spinned. What is the probability of getting 12 or 35?   
# P(1/38 or 1/38) = P(1/38) + P(1/38) = P(1/19) = P(0.05263157894)
k = 0
n = 10000000
for i in range(0, n):
    x = np.random.randint(0, 38)
    if (x == 12) or (x == 35):
        k = k + 1
print(k / n)
```

1) Дополните код Монте-Карло последовательности независимых испытаний расчетом соответствующих вероятностей (через биномиальное распределение) и сравните результаты.  
2) Повторите расчеты биномиальных коэффициентов и вероятностей k успехов в последовательности из n независимых испытаний, взяв другие значения n и k.

```python
# Вероятность того, что в 4-х испытаниях орел выпадет ровно два раза
k, n = 0, 10000000
a = np.random.randint(0, 2, n)
b = np.random.randint(0, 2, n)
c = np.random.randint(0, 2, n)
d = np.random.randint(0, 2, n)
x = a + b + c + d
for i in range(0, n):
    if x[i] == 2:
        k = k + 1
print(k / n)
```

```python
# Вероятность того, что в 4-х испытаниях орел выпадет ровно два раза
p = math.factorial(4) / (math.factorial(2) * math.factorial(2)) * 0.5**2 * 0.5**2
print(p)
```

```python
# Вероятность того, что в 5-ти испытаниях орел выпадет ровно 3 раза
p = math.factorial(5) / (math.factorial(2) * math.factorial(3)) * 0.5**3 * 0.5**2
print(p)
```
